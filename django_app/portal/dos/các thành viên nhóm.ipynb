## 📘 Project Overview (Đề tài)

Dự án tập trung nghiên cứu và triển khai **Hệ Thống Điều Phối Cứu Hộ Thông Minh**, ứng dụng thuật toán AI cơ bản để xác định vị trí, quản lý yêu cầu cứu hộ và lựa chọn trạm cứu hộ phù hợp.


### ✨ Core Features

* **Location Detection:** Xác định vị trí người gặp nạn bằng GPS, chọn trực tiếp trên bản đồ hoặc nhập địa chỉ ô 3 từ.
* **Rescue Request Management:** Tiếp nhận, lưu trữ và quản lý yêu cầu cứu hộ theo mức độ khẩn cấp và trạng thái xử lý.
* **Smart Dispatching:** Sử dụng thuật toán A* kết hợp dữ liệu OSRM để lựa chọn trạm cứu hộ phù hợp và tìm tuyến đường tối ưu.
* **Station Management:** Quản lý thông tin trạm cứu hộ, trạng thái hoạt động và số lượng phương tiện còn sẵn sàng.
* **Route Visualization:** Hiển thị vị trí trạm, vị trí người gặp nạn và tuyến đường cứu hộ trên bản đồ.
* **Status Tracking:** Theo dõi yêu cầu từ lúc chờ tiếp nhận, đã phân công, đang di chuyển đến khi hoàn thành hoặc hủy.
